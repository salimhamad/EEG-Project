```python
import torch
import torch.nn as nn
import torch.optim as optim

# ١. دیاریکردنی Hyperparameters (ئەو دوگمانەی دەستکارییان دەکەین)
LEARNING_RATE = 0.001  # خێرایی فێربوون
BATCH_SIZE = 32        # مۆدێلەکە ٣٢ پارچە داتا بەیەکەوە دەخوێنێت
EPOCHS = 10            # مۆدێلەکە ١٠ جار هەموو داتاکان دووبارە دەکاتەوە
HIDDEN_UNITS = 128     # ژمارەی "دەمارەکان" لە ناو مۆدێلەکەدا

# ٢. دروستکردنی مۆدێلەکە
model = nn.Sequential(
    nn.Linear(250, HIDDEN_UNITS), # ٢٥٠ داتای EEG وەردەگرێت
    nn.ReLU(),
    nn.Linear(HIDDEN_UNITS, 2)    # دەیکاتە ٢ وشە
)

# ٣. ئامرازەکان بۆ فێربوون
# لێرەدا Learning Rate بەکاردێت
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
criterion = nn.CrossEntropyLoss()

# ٤. خولی مەشقپێکردن (Training Loop)
for epoch in range(EPOCHS):
    # لێرەدا BATCH_SIZE کاری خۆی دەکات (داتاکە پارچە پارچە دەکرێت)
    # خەیاڵ بکە داتاکەمان لێرەدا بە پارچەی ٣٢ تایی دەڕوات

    print(f"قۆناغی (Epoch) ژمارە {epoch+1} دەستی پێکرد...")

    # ... لێرەدا کۆدی مەشقەکە دەبێت ...
```

```python
from torch.utils.data import DataLoader, TensorDataset

# گریمانە دەکەین ئەمە داتای ناو کاگڵەکەتە:
# X: شەپۆلەکان (Features)، Y: وشەکان (Labels)
# لێرەدا بە دەستکرد دروستی دەکەین بۆ تاقیکردنەوە
X = torch.randn(1000, 250) # ١٠٠٠ نموونەی EEG
Y = torch.randint(0, 2, (1000,)) # ١٠٠٠ وشە (٠ یان ١)

# دروستکردنی داتاسێت و لۆدەر
dataset = TensorDataset(X, Y)
train_loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)

# ٤. خولی مەشقپێکردن (Training Loop) بە شێوەی تەواو
for epoch in range(EPOCHS):
    running_loss = 0.0

    # لێرەدا DataLoader داتاکە پارچە پارچە دەکات و دەیدات بە مۆدێلەکە
    for batch_idx, (signals, labels) in enumerate(train_loader):

        # هەنگاوی ١: پاککردنەوەی گادیێنت
        optimizer.zero_grad()

        # هەنگاوی ٢: پێشبینیکردن (Forward Pass)
        outputs = model(signals)

        # هەنگاوی ٣: هەژمارکردنی هەڵە (Loss)
        loss = criterion(outputs, labels)

        # هەنگاوی ٤: گەڕانەوە بۆ دواوە و فێربوون (Backpropagation)
        loss.backward()

        # هەنگاوی ٥: نوێکردنەوەی کێشی دەمارەکان (Update Weights)
        optimizer.step()

        running_loss += loss.item()

    print(f"قۆناغی [{epoch+1}/{EPOCHS}] - تێکڕای هەڵە: {running_loss/len(train_loader):.4f}")
```

٢. ئەم کۆدە چۆن "ئیش دەکات"؟ (قۆناغ بە قۆناغ)
بۆ ئەوەی تێبگەیت لە ناو مۆدێلەکەدا چی دەگوزەرێت، ئەم وێنەیە خەیاڵ بکە:
DataLoader (دابەشکەر): لە کۆگای داتاکان (X, Y)، هەر جارەی ٣٢ دانە هەڵدەگرێت و دەیبات بۆ ناو مۆدێلەکە. بۆچی ٣٢؟ چونکە ئەگەر هەموو ١٠٠٠ دانەکە بەیەکەوە بەرین، مێشکی کۆمپیوتەرەکە (RAM) قورس دەبێت و زانیارییەکان تێکەڵ دەبن.
Forward Pass (چوونە پێش): شەپۆلەکانی EEG دەچنە ناو چینەکانی مۆدێلەکە (nn.Linear). لێرەدا ژمارەکان لێکدران و کۆکردنەوەیان بۆ دەکرێت تا دەگەنە کۆتایی و مۆدێلەکە دەڵێت: "بە ئەگەری ٨٠٪ ئەمە وشەی 'ئاو'ە".
Loss Function (پێوەری هەڵە): مۆدێلەکە سەیری وشە ڕاستەقینەکە دەکات. ئەگەر هەڵەی کردبێت، ژمارەیەکی گەورە دروست دەبێت (Loss). ئەگەر ڕاست بووبێت، ژمارەکە بچووک دەبێت.
Backpropagation (فێربوون): ئەمە "سیحرەکەیە". مۆدێلەکە بە دوای ئەو "کێشانەدا" (Weights) دەگەڕێت کە بوونەتە هۆی هەڵەکە و کەمێک دەستکارییان دەکات. لێرەدا Learning Rate دیاری دەکات ئەو دەستکارییە چەندێک گەورە بێت.
Optimizer (ڕێکخەر): وەک ئەندازیارێک وایە، هەموو پارچەکان نوێ دەکاتەوە بۆ ئەوەی لە خولی داهاتوودا (Batch ی دواتر) هەمان هەڵە نەکەنەوە.

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import matplotlib.pyplot as plt
import numpy as np

# ١. دروستکردنی داتای دەستکرد بۆ تاقیکردنەوە (Simulation)
def create_data(samples=100):
    # دروستکردنی شەپۆلی EEG (وەک ئەوەی لە مێشکەوە هاتبێت)
    x = torch.randn(samples, 250)
    # دروستکردنی لیبڵ (وشەی ٠ یان ١)
    y = torch.randint(0, 2, (samples,))
    return x, y

X_data, Y_data = create_data(200)

# ٢. پیشاندانی داتاکە (ئەمە ئەو بەشەیە کە داوات کردبوو)
def plot_data(x, y, index=0):
    plt.figure(figsize=(10, 4))
    plt.plot(x[index].numpy(), color='blue', label=f'Signal of Word: {y[index].item()}')
    plt.title(f"EEG Signal Sample - Word Category: {y[index].item()}")
    plt.xlabel("Time points")
    plt.ylabel("Microvolts")
    plt.legend()
    plt.show()

# پیشاندانی یەکەم نموونە پێش مەشقپێکردن
plot_data(X_data, Y_data, index=5)

# ٣. ڕێکخستنی Hyperparameters
LEARNING_RATE = 0.01
BATCH_SIZE = 16
EPOCHS = 20

# ئامادەکردنی DataLoader
dataset = TensorDataset(X_data, Y_data)
train_loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)

# ٤. دروستکردنی مۆدێل
model = nn.Sequential(
    nn.Linear(250, 64),
    nn.ReLU(),
    nn.Linear(64, 2)
)

optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
criterion = nn.CrossEntropyLoss()

# ٥. خولی مەشقپێکردن و پیشاندانی ئەنجام
print("دەستپێکردنی مەشقپێکردن...")
losses = []

for epoch in range(EPOCHS):
    epoch_loss = 0
    for batch_x, batch_y in train_loader:
        optimizer.zero_grad()
        outputs = model(batch_x)
        loss = criterion(outputs, batch_y)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()

    avg_loss = epoch_loss / len(train_loader)
    losses.append(avg_loss)
    if (epoch+1) % 5 == 0:
        print(f"Epoch [{epoch+1}/{EPOCHS}], Loss: {avg_loss:.4f}")

# ٦. پیشاندانی گرافی فێربوون (Training Curve)
plt.plot(losses)
plt.title("Learning Progress (Loss over Time)")
plt.xlabel("Epoch")
plt.ylabel("Error (Loss)")
plt.show()
```

ئەم کۆدە چی نوێی تێدایە؟
plot_data: ئەم فانکشنە ڕێگەت پێ دەدات یەکێک لە شەپۆلەکانی ناو داتاسێتەکە ببینی. لە کاگڵ، هەمیشە ئەمە بکە بۆ ئەوەی بزانیت ئایا داتاکە "پاکە" یان نا.
Tracking Losses: لیستێکی مۆ پێکهێناوە بە ناوی losses. ئەمە وەک "نمرەی مۆدێلەکەیە" لە کاتی مەشقەکەدا.
Loss Graph: لە کۆتاییدا گرافێکت بۆ دروست دەکات. ئەگەر هێڵەکە بەرەو خوارەوە بڕوات، واتا مۆدێلەکەت فێر دەبێت. ئەگەر هێڵەکە بە بەرزی مایەوە یان بەرز بووەوە، واتا Learning Rateەکەت هەڵەیە و دەبێت دەستکاری بکەیت.
چۆن لە کاگڵ داتای ڕاستەقینە نیشان بدەیت؟
لەو کێبڕکێیەی کە بەشداریت تێدا کردووە، لە جیاتی torch.randn (کە داتای هەڕەمەکییە)، دەبێت فایلەکان (زۆربەی کات .csv یان .npy) بخوێنیتەوە بەم شێوەیە: data = pd.read_csv('train.csv') دواتر بە هەمان فانکشنی plt.plot دەتوانیت شەپۆلی مێشکە ڕاستەقینەکە ببینی.

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# ١. وەبیرهێنانەوەی وشەکان
classes = ["ئاو (Water)", "نان (Bread)"]

# ٢. تاقیکردنەوەی مۆدێلەکە لەسەر هەموو داتاکان
model.eval() # مۆدێلەکە دەخەینە دۆخی تاقیکردنەوە
all_preds = []
all_labels = []

with torch.no_grad():
    for signals, labels in train_loader:
        outputs = model(signals)
        _, predicted = torch.max(outputs, 1)
        all_preds.extend(predicted.numpy())
        all_labels.extend(labels.numpy())

# ٣. دروستکردنی ماتریسەکە
cm = confusion_matrix(all_labels, all_preds)

# ٤. کێشانی گرافەکە
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix: کێشەکانی مۆدێلەکە لێرە دەردەکەون")
plt.show()
```

چۆن ئەم گرافە بخوێنیتەوە؟
هێڵە لارییە ڕاستەکە (Diagonal): ئەگەر ژمارەکانی ناو ئەم هێڵە زۆر بن، واتا مۆدێلەکەت زۆر باشە. واتە "ئاو"ی وەک "ئاو" ناسیوەتەوە.
خانەکانی تر: ئەگەر لە خانەیەکدا ژمارەیەک ببینی، واتا مۆدێلەکە لەوێدا "تەمەڵە". بۆ نموونە ئەگەر مۆدێلەکە ٣٠ جار وشەی "نان"ی وەک "ئاو" بینیبێت، لێرەدا دەردەکەوێت.
سوودی ئەمە بۆ کاگڵ چییە؟
لە کێبڕکێیەکدا وەک ئەوەی تۆ کردووتە، لەوانەیە ١٠ جۆر وشە هەبێت. کاتێک Confusion Matrix دروست دەکەیت، دەبینی کە مۆدێلەکەت لە ٨ وشەدا زۆر زیرەکە، بەڵام ٢ وشە هەن هەمیشە تێکەڵی دەەکات (بۆ نموونە وشەی "سیێ" و "چوار" چونکە لە لەرەلەردا لێک نێزیکن).
